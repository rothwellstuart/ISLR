```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
college = pd.read_csv("College.csv")

college.rename(columns={college.columns[0]:'College'}, inplace=True)

college.shape
college.dtypes
college.head()
college.tail()
```

```python
# Summary stats

college.describe()
```

```python
# Median

college.median()
```

```python
# Scatter plots

# plot_cols = ['Private' 'Apps', 'Accept', 'Enroll', 'Top10perc', 'Top25perc', 'F.Undergrad', 'P.Undergrad', 'Outstate', 'Room.Board']

pd.plotting.scatter_matrix(college.iloc[:, 2:12], figsize=[12, 12])
```

```python
corr_matrix = college.iloc[:,2:12].corr()
corr_names = college.iloc[:,2:12].columns.values

corr_names

len(list(corr_names))
```

```python
fig = plt.figure(figsize=(12, 12))
ax = plt.gca()
ax.set_xticks(np.arange(len(list(corr_names))))
ax.set_xticklabels(list(corr_names))
ax.set_yticks(np.arange(len(list(corr_names))))
ax.set_yticklabels(list(corr_names))
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax.yaxis.set_major_locator(ticker.MultipleLocator(1))

# Rotate and align top ticklabels
plt.setp([tick.label2 for tick in ax.xaxis.get_major_ticks()], rotation=45, ha="left", va="center",rotation_mode="anchor")

im = ax.matshow(corr_matrix, interpolation='none')
fig.colorbar(im)
fig.tight_layout()
plt.show()
```

```python
# Boxplot: Outstate by Private

college.loc[:,['Private', 'Outstate']].boxplot(column='Outstate', by='Private')
```

```python
### Pick up from PDF page 69, (c) iv.

# Create a new qualitative variable, called Elite, by binning
# the Top10perc variable. We are going to divide universities
# into two groups based on whether or not the proportion
# of students coming from the top 10 % of their high school
# classes exceeds 50 %

college['Elite'] = np.where(college['Top10perc']>50, 'Elite', 'Not Elite')

college.head()
```

```python
college[['Elite']].describe()
```

```python
# Boxplot: Outstate by Elite

college.loc[:,['Elite', 'Outstate']].boxplot(column='Outstate', by='Elite')
```
